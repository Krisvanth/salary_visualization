# **Visualization**

```python
# selecting rows based on condition
year_2020 = job[job['work_year']==2020]
title1=year_2020['job_title']
salary1=year_2020['salary']
```

```python
# selecting rows based on condition
year_2021 = job[job['work_year']==2021]
title2=year_2021['job_title']
salary2=year_2021['salary']
```

```python
# selecting rows based on condition
year_2022 = job[job['work_year']==2022]
title3=year_2022['job_title']
salary3=year_2022['salary']
```

```python
plt.figure(figsize= (10,10))
plt.barh(title1,salary1,color='red')
plt.xlabel('Salary')
plt.ylabel('Job Title')
plt.title("2020 Salary Trend",fontsize=20)
plt.show()
```

```python
plt.figure(figsize= (10,13))
plt.barh(title2,salary2,color='b')
plt.xlabel('Salary')
plt.ylabel('Job Title')
plt.title("2021 Salary Trend",fontsize=20)
plt.show()
```

```python
plt.figure(figsize= (10,13))
plt.barh(title3,salary3,color='g')
plt.xlabel('Salary')
plt.ylabel('Job Title')
plt.title("2022 Salary Trend",fontsize=20)
plt.show()
```
